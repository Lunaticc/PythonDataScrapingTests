```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
from urllib.request import urlopen

class GetPage:
    def __init__(self, url):
        self.url_ = url
        self.webpage_ = self.get_webpage()
        self.title_ = self.get_pageTitle()
        
    def get_webpage(self):
        webpage = urlopen(self.url_).read()
        return bs(webpage, 'lxml')

    def get_pageTitle(self):
        regfix = re.compile('[\W\s]+')
        return re.sub(regfix, '_', self.webpage_.find('title').string)
        
    def get_Tag(self, tagType = None, tagClass = None):
        self.title_ = self.title_ + tagType
        if tagType == None:
            print('Remember A tagType, please')
        if tagType == 'table':
            if tagClass == None:
                self.table_ = self.webpage_.find('table')
            else:
                self.table_ = self.webpage_.find('table', class_=tagClass)
            self.title_ = self.title_ + '_{tagtype}'.format(tagtype=tagType)
            return self.table_
        if tagType == 'tables':
            self.tables_ = self.webpage_.find_all('table')
            self.title_ = self.title_ + '_{tagtype}'.format(tagtype=tagType)
            return self.tables_
        
    def get_rows(self, table = None):
        if table == None:
            self.rows_ = self.table_.find_all('tr')
            return self.rows_
        else:
            self.rows_ = table.find_all('tr')

    
    def make_df(self, rows = None):
        word = re.compile('t[hd]')
        if rows == None:
            first = self.rows_[0]
            second = self.rows_[1:]
        else:
            first = rows[0]
            second = rows[1:]
        headers_ = [item.text.replace('\n', '') for item in first.find_all('th')]
        values_ = [[item.text.replace('\n', '') for item in item.find_all(word)] for item in second]
        self.df_ = pd.DataFrame(data = values_, columns = headers_)
        
    def correct_df(self, df=None):
        x = 0
        if(df == None):
            while(x<len(self.df_.iloc[:,-1:])):
                if('None' in str(self.df_.iloc[x,-1:])):
                    self.df_.loc[x] = self.df_.loc[x].shift(1, axis=0, fill_value='-')
                x=x+1
            return self.df_
        else:
            while(x<len(df.iloc[:,-1:])):
                if('None' in str(df.iloc[x,-1:])):
                    df.loc[x] = df.loc[x].shift(1, axis=0, fill_value='-')
                x=x+1
            return df
    
    def to_csv(self):
        self.df_.to_csv(str(self.title_)+'.csv')
        return '.csvfile '+ self.title_ +' created...'
    
    def to_excel(self):
        self.df_.to_excel(str(self.title_)+'.xlsx')
        return '.xlsxFile '+self.title_+' creater...'
```
